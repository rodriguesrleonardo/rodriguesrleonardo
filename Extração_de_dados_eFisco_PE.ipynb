```python
import os
import re
import csv
import pdfplumber

# Nome atualizado da pasta virtual na nuvem
PASTA_PDFS = "./meus_pdfs_27197888026036"
ARQUIVO_SAIDA = "extratos_efisco_27197888026036_consolidados.csv"

# Expressões regulares para capturar os dados do eFisco
regex_grupo = re.compile(r"GRUPO DE MERCADORIA:\s*(.*?)\s*([\d\.,]+)$")
regex_nota = re.compile(r"^(\d{3}\.\d\.\d{9}-\d)\s+(\d+)\s+([I|N])\s+([A-Z]{2})\s+(\d{14})\s+(\d+)\s+([\d\.,]+)$")

dados_consolidados = []

if not os.path.exists(PASTA_PDFS):
    print(f"ERRO: A pasta '{PASTA_PDFS}' não foi encontrada na barra lateral.")
    print("Verifique se o nome está idêntico (letras maiúsculas/minúsculas e números) ao que você criou.")
    exit()

print(f"Iniciando a extração dos PDFs carregados na pasta: {PASTA_PDFS}...\n")

for arquivo_nome in os.listdir(PASTA_PDFS):
    if arquivo_nome.lower().endswith(".pdf"):
        caminho_completo = os.path.join(PASTA_PDFS, arquivo_nome)
        print(f"Processando: {arquivo_nome}")

        grupo_atual = "Não Identificado"

        try:
            with pdfplumber.open(caminho_completo) as pdf:
                for pagina in pdf.pages:
                    texto = pagina.extract_text()
                    if not texto:
                        continue

                    for linha in texto.split("\n"):
                        linha = linha.strip()

                        match_grupo = regex_grupo.search(linha)
                        if match_grupo:
                            grupo_atual = match_grupo.group(1).strip()
                            continue

                        match_nota = regex_nota.search(linha)
                        if match_nota:
                            reg_nota, cod_prod, origem, uf, cnpj, num_nf, icms_devido = match_nota.groups()

                            dados_consolidados.append({
                                "Nome do Arquivo": arquivo_nome,
                                "Grupo Mercadoria": grupo_atual,
                                "Registro da Nota": reg_nota,
                                "Código Produto": cod_prod,
                                "Origem": origem,
                                "UF": uf,
                                "CNPJ Emitente": cnpj,
                                "Nota Fiscal": num_nf,
                                "ICMS Devido": icms_devido
                            })
        except Exception as e:
            print(f"Erro no arquivo {arquivo_nome}: {e}")

if dados_consolidados:
    with open(ARQUIVO_SAIDA, "w", newline="", encoding="utf-8-sig") as csv_file:
        campos = ["Nome do Arquivo", "Grupo Mercadoria", "Registro da Nota", "Código Produto", "Origem", "UF", "CNPJ Emitente", "Nota Fiscal", "ICMS Devido"]
        writer = csv.DictWriter(csv_file, fieldnames=campos, delimiter=";")
        writer.writeheader()
        writer.writerows(dados_consolidados)

    print("\n--- PROCESSAMENTO CONCLUÍDO ---")
    print(f"Sucesso! {len(dados_consolidados)} linhas extraídas.")
    print(f"O arquivo '{ARQUIVO_SAIDA}' foi gerado com sucesso na barra lateral.")
else:
    print("\nNenhuma nota fiscal encontrada. Verifique se os arquivos PDF foram arrastados corretamente para dentro da pasta.")
```
